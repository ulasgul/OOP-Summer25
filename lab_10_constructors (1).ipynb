```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        print(f"Car created: {brand} {model}")
```

```python
car1 = Car ("Toyota", "Corolla")
car2 = Car ("Ford", "Mustang")
car3 = Car ("Mercedes", "C200")
```
